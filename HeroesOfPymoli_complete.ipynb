```python
# Retrieve the Item ID, Item Name, and Item Price columns
item_breakdown = purchase_data[["Item ID", "Item Name", "Price"]]

# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
item_count = item_breakdown.groupby(["Item ID","Item Name"])["Price"].count()
item_price = item_breakdown.groupby(["Item ID","Item Name"])["Price"].mean()
item_total = item_breakdown.groupby(["Item ID","Item Name"])["Price"].sum()
# Create a summary data frame to hold the results
item_df = pd.DataFrame(item_count)
item_df["Item Price"] = item_price
item_df["Total Purchase Value"] = item_total
item_df.columns = ["Purchase Count","Item Price","Total Purchase Value"]
# Sort the purchase count column in descending order
most_pop = item_df.sort_values("Purchase Count", ascending = False)
most_pop.head()


# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame
```
